```python
import numpy as np
from numpy.core.fromnumeric import reshape
```

numpy ndarray object

```python
a = np.array([1,2,3])
print(a)
```

```python
a1 = np.array([[1,2,3],[4,5,6]])
print("\n",a1)
```

ndnimm parameter

```python
a2 = np.array([1,2,3,4,5,6],ndmin=2)
print(a2)
```

dtype parameter 

```python
a3 = np.array([1,2,3],dtype=complex) 
print(a3)
```

ndarray.shape attribute<br>
This array attribute returns a tuple consisting of array dimensions. It can also be used to resize the array.

```python
print(a1.shape)
```

reshape

```python
a1.shape=(3,2)
print(a1)
```

```python
b1 = a1.reshape(3,2)
print(b1)
```

ndarray.ndim<br>
This array attribute returns the number of array dimensions.

```python
print("\n",a1.ndim)
```

numpy.itemsize<br>
This array attribute returns the length of each element of array in bytes.

```python
print(a1.itemsize)
```

numpy.flags<br>
The ndarray object has the following attributes. Its current values are returned by this function.

```python
x = np.array([1,2,3,4,5]) 
print(x.flags)
```

numpy.empty<br>
It creates an uninitialized array of specified shape and dtype. It uses the following constructor âˆ’

```python
x1 = np.empty([3,2], dtype = int) 
print(x1)
```

```python
x2 = np.ones([2,2], dtype = int) 
print("\n",x2)
```

numpy.asarray<br>
This function is similar to numpy.array except for the fact that it has fewer parameters. <br>
This routine is useful for converting Python sequence into ndarray.

```python
x3 = [1,2,3]
```

```python
print("\n",np.asarray(x3))
```

```python
x4 = (1,2,3)
print("\n",np.asarray(x4))
```

numpy.frombuffer<br>
This function interprets a buffer as one-dimensional array. <br>
Any object that exposes the buffer interface is used as parameter to return an ndarray.

```python
s = b'Hello World' 
print("\n",np.frombuffer(s,dtype='S1'))
```

numpy.fromiter<br>
This function builds an ndarray object from any iterable object. <br>
A new one-dimensional array is returned by this function.

```python
list= range(1,5)
print(list)
```

```python
it = iter(list)  
```

```python
print(np.fromiter(it, dtype = float))
```

numpy.arange<br>
This function returns an ndarray object containing evenly spaced values within a given range. <br>
The format of the function is as follows âˆ’

numpy.arange(start, stop, step, dtype)

```python
x5 = np.arange(10,50,2,dtype=int)
```

```python
print(x5)
```

numpy.linspace<br>
This function is similar to arange() function. In this function, instead of step size, <br>
the number of evenly spaced values between the interval is specified. The usage of this function is as follows âˆ’

```python
x6  = np.linspace(10,20,5)
print(x6)
```

```python
x7 = np.linspace(10,20,5,endpoint=False,retstep=True)
print(x7)
```

numpy.logspace<br>
This function returns an ndarray object that contains the numbers that are evenly spaced on a log scale. <br>
Start and stop endpoints of the scale are indices of the base, usually 10.

numpy.logspace(start, stop, num, endpoint, base, dtype)

```python
x8 = np.logspace(1.0, 2.0, num = 10) 
print(x8)
```

```python
x9 = np.logspace(1,10,num=10,base=2)
print(x9)
```
